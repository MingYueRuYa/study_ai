```python
import pandas as pd
df_ads = pd.read_csv('易速鲜花微信软文.csv') # 读入数据
# df_ads.astype(int)
df_ads.head(10) # 显示前几行数据
```

```python
df_ads.isna().sum()
```

```python
df_ads = df_ads.dropna() 
```

```python
X = df_ads.drop(['浏览量'],axis=1) # 特征集，Drop掉便签相关字段
y = df_ads.浏览量 # 标签集
```

```python
X.head() # 显示前几行数据
```

```python
y.head() # 显示前几行数据
```

```python
# 将数据集进行80%（训练集）和20%（验证集）的分割
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                   test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train) # 用训练集数据，训练机器，拟合函数，确定参数
```

```python
y_pred = model.predict(X_test) # 预测测试集的Y值
```

```python
df_ads_pred= X_test.copy()#测试集特征数据
df_ads_pred['浏览量真值'] =y_test
df_ads_pred['浏览量预测值'] = y_pred
df_ads_pred
```

```python
print("线性回归预测集评分：", model.score(X_test, y_test))
print("训练集评分：", model.score(X_train, y_train))
```

```python
print('当前模型的4个特征的权重分别是：', model.coef_)
print('当前模型的截距(偏振)是：', model.intercept_)
```
